# Line plot

```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5]
y = [0,3,2,22,22,10]
plt.plot(x,y)
plt.xlabel("days")
plt.ylabel("counts")
plt.show()
```

# Line plot Axis controls

```python
import numpy as np
x = np.arange(0,400)
y = np.cos(x)
plt.plot(x,y)
plt.xlim(300,400)
plt.show()
```

# Labels and Legends

```python
x = [0,1,2,3,4,5]
y1 = [0,3,2,22,22,10]
y2 =[1,7,5,4,3,8]
plt.plot(x,y1,c="cyan",label="location1")
plt.plot(x,y2,c="black",label="location2")
plt.xlabel("days")
plt.ylabel("counts")
plt.legend(loc="lower right")
plt.show()
```

# Bar chart

```python
x= ["Mon","Tues","Wed","Thurs","Fri","Sat","Sun"]
item_1_sales = [1,2,3,4,5,6,7]
plt.bar(x,item_1_sales,label ="item1")
plt.xlabel("days")
plt.ylabel("counts")
plt.savefig("bar_chart.png",dpi=500)
plt.legend()
plt.show()
```

```python
x= ["Mon","Tues","Wed","Thurs","Fri","Sat","Sun"]
item_1_sales = [1,2,3,4,5,6,7]
item_2_sales = [2,4,6,5,6,7,9]
plt.bar(x,item_1_sales,color="r",label ="item1")
plt.bar(x,item_2_sales,color="g",alpha=0.2,label ="item2")
plt.xlabel("days")
plt.ylabel("counts")
plt.savefig("bar_chart.png",dpi=200)
plt.legend(loc = "upper left")
plt.legend()
plt.show()
```

# Histogram

```python
import pandas as pd
data = pd.read_csv("sample_data/auto-mpg.csv")
data.head()
```

```python
plt.hist(data["mpg"],color="green",rwidth=0.8,density=True)
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Density")
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(10,3.5))
plt.scatter(data["weight"],data["mpg"],c="k",marker="+")
plt.title("MPG vs weight")
plt.xlabel("weight")
plt.ylabel("mpg")
plt.show()
```

# Sub-Figures

```python
plt.figure(figsize=[8,6])
plt.subplot(2,1,1)
plt.scatter(data["weight"],data["mpg"],c="k")
plt.ylabel("MPG")
plt.xlabel("Weight")
plt.subplot(2,1,2)
plt.scatter(data["displacement"],data["mpg"],c="blue")
plt.ylabel("MPG")
plt.xlabel("Displacement")
plt.show()
```
